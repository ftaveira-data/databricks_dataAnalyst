```
import pyspark.pandas as ps

file_location = "/FileStore/tables/analise_dados/prouni_2021/relatorio_inscricao_dados_abertos_fies_22021.csv"
delimiter = ";" 

# Lendo o arquivo CSV
df_prouni_21 = ps.read_csv(file_location, header=0, sep=delimiter)

# Exibindo as primeiras linhas do DataFrame
df_prouni_21.head()
```

```
df_prouni_21.columns
```

```
# Verifique se a coluna 'ID do estudante' existe
if 'ID do estudante' in df_prouni_21.columns:
    # Renomeie a coluna
    df_prouni_21 = df_prouni_21.rename(columns={'ID do estudante':'id_do_estudante'})
    # Defina a coluna renomeada como índice
    df_prouni_21.set_index('id_do_estudante', inplace=True)
else:
    print("A coluna 'ID do estudante' não existe no DataFrame.")
```

```
df_prouni_21.head()
```

```
df_prouni_21_v2 = df_prouni_21.dropna()
```

```
df_prouni_21_v2.head()
```

```
# Agrupar por 'sexo' e contar as ocorrências
count_sex = df_prouni_21_v2.groupby("Sexo").size()

print(count_sex)
```

```
# Calcular a proporção
total_count = len(df_prouni_21_v2)

print(total_count)
```

```

proportion_sex = count_sex / total_count
print(proportion_sex)
```

```
# Exemplo de uso de query para filtrar e depois contar
count_m = df_prouni_21_v2.query("sexo == 'M'").shape[0]
count_f = df_prouni_21_v2.query("sexo == 'F'").shape[0]

print("Homens:", count_m)
print("Mulheres:", count_f)
```

```
total_por_sexo = ps.sql('''
                         SELECT Sexo, count(*) as total_genero from {DF}                                                
                         Group By Sexo
                         Order By total_genero DESC
                         ''',DF=df_prouni_21_v2)

total_por_sexo.head()
```

```
total_por_sexo.plot.bar(x="Sexo", y="total_genero", title="Distribuição por Sexo")
```

```
disciplinas=['Redação',    'Matemática e suas Tecnologias',    'Linguagens, Códigos e suas Tec',    'Ciências Natureza e suas Tec',    'Ciências Humanas e suas Tec']

df_five_columns = df_prouni_21_v2[disciplinas]
```

```
print(df_five_columns)
```

```
df_five_columns['Redação'] = df_five_columns['Redação'].apply(
    lambda x: float(x.replace(',', '.')) if isinstance(x, str) else x)

df_five_columns['Matemática e suas Tecnologias'] = df_five_columns['Matemática e suas Tecnologias'].apply(
    lambda x: float(x.replace(',', '.')) if isinstance(x, str) else x)

df_five_columns['Linguagens, Códigos e suas Tec'] = df_five_columns['Linguagens, Códigos e suas Tec'].apply(
    lambda x: float(x.replace(',', '.')) if isinstance(x, str) else x)

df_five_columns['Ciências Natureza e suas Tec'] = df_five_columns['Ciências Natureza e suas Tec'].apply(
    lambda x: float(x.replace(',', '.')) if isinstance(x, str) else x)

df_five_columns['Ciências Humanas e suas Tec'] = df_five_columns['Ciências Humanas e suas Tec'].apply(
    lambda x: float(x.replace(',', '.')) if isinstance(x, str) else x)

```
